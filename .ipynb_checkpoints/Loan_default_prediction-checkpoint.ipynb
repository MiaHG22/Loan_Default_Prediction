```python
# Now we need to transform the categorical features to be numeric before moving on to modeling
le=LabelEncoder()
onehot_encoder=OneHotEncoder(sparse=False)
df['married']=le.fit_transform(df['married'])
df['house_ownership']= le.fit_transform(df['house_ownership'])
df['car_ownership']= le.fit_transform(df['car_ownership'])
```

```python
high_card_features = ['profession', 'city', 'state']
count_encoder=ce.CountEncoder()
count_encoded = count_encoder.fit_transform(df[high_card_features])
df=df.join(count_encoded.add_suffix("_count"))
```

```python
df=df.drop(labels=['profession', 'city', 'state'], axis=1)
```

```python
df.head()
```

```python
# Now we can measure the performance of machine learning models 
X=df.drop('risk_flag', axis=1)
y=df['risk_flag']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=78)
```

```python
# Determine the shape of our training 
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Resample to balance the dataset
rf_clf=RandomForestClassifier(criterion='gini', bootstrap=True, random_state=100)
smote_sampler = SMOTE(random_state=9)
pipeline= Pipeline(steps= [['smote', smote_sampler], ['classifier', rf_clf]])
rf_clf.fit(X_train, y_train)
y_pred=pipeline.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score, accuracy_score
print('______________________TEST SCORES______________________')
print(f'Recall: {round(recall_score(y_test, y_pred)*100, 4)}')
print(f'Precision: {round(precision_score(y_test, y_pred)*100, 4)}')
print(f'F1-score: {round(f1_score(y_test, y_pred)*100, 4)}')  
print(f'Accuracy score: {round(accuracy_score(y_test, y_pred)*100, 4)}')
```
